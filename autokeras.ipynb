```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import autokeras as ak
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/data.csv')

# print(df.head())
```

```python
df.head()
```

```python
df=df.drop('timestamp', axis=1)
df.head()
```

```python
print(df.shape)
```

```python
X = df.drop("label",axis=1)
y = df['label']
```

```python
X_train, X_test_check, y_train, y_test_check = train_test_split(
    X, y, test_size=0.35, random_state=42)
```

```python
X_train.shape
```

```python
X_test, X_validate, y_test, y_validate = train_test_split(
    X_test_check, y_test_check, test_size=2/7, random_state=42)
```

```python
df_validate = X_validate
```

```python
y_validate.head()
```

```python
df_validate.head()
```

```python
df_validate = pd.concat([df_validate, y_validate], axis=1)
df_validate.head()
```

```python
df_validate.to_csv('validate.csv',index=False)
```

```python
clf = ak.StructuredDataClassifier(max_trials=20)
clf.fit(X_train, y_train, epochs=50)
```

```python
loss, acc = clf.evaluate(X_test, y_test, verbose=0)
print('Accuracy: %.5f' % acc)
```

```python
model = clf.export_model()
model.summary()
```

```python
y_pred = model.predict(X_test)
```
